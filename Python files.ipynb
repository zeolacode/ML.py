# Introdução ao JSON 

- JavaScript Object Notation (JSON) é um formato de arquivo padrão que armazena dados em um formato hierárquico semelhante a eXtensible Markup Language (XML). JSON nada mais é do que uma hierarquia de listas e dicionários. Os programadores referem-se a este tipo de dados como dados semiestruturados ou dados hierárquicos.

```python
# exemplo de um json de um objeto que descreve uma pessoa!

var_agenda = {
    "nome": "Jose Carlos",
    "sobrenome": "Macoratti",
    "idade": 45,
    "endereco":  {  
                           "endereco": "Rua Projetada, 200", 
                           "cidade": "Santos", 
                           "estado": "SP",
                           "cep": 11054058
                       },
    "telefone": [
                            {
                                "tipo": "fixo",
                                "numero": "212 555-1234"
                            },
                            {
                                "tipo": "celular",
                                "numero": "646 555-4567"
                            }
                     ]
}
```

- O arquivo acima pode ser parecido com o código Python. Você pode ver colchetes que definem dicionários e colchetes que definem listas. JSON exige que haja um único elemento raiz. Uma lista ou dicionário pode cumprir essa função. JSON requer aspas duplas para incluir strings e nomes. As aspas simples não são permitidas em JSON.

- Os arquivos JSON são sempre sintaxe JavaScript legal. JSON também é geralmente válido como código Python, conforme demonstrado pelo seguinte programa Python.

```python
var_agenda.keys()
```

```python
var_agenda.values()
```

```python
var_agenda['telefone']
```

```python
var_agenda['telefone'][0]
```

```python
var_agenda.items()
```

## Manipulando json

```python
import json 
```

```python
json_name = '{"Primeiro":"Pedro", "Segundo":"Zeola"}'
obj = json.loads(json_name)

print(f"Primeiro nome: {obj['Primeiro']}")
print(f"Segundo nome: {obj['Segundo']}")
```

```python
import requests

# realizando um resquest em um json que está em um repositório do meu GitHub 

r = requests.get("https://raw.githubusercontent.com/zeolacode/Data/master/JSON.json")

print(r.json())
```

```python
# Gerando strings JSON a partir de um dicionário
frutas  = {0 : "banana", 1 : "goiaba"}
print(json.dumps(frutas))
```

# Image
- Para trabalhar com visão computacional, o programador Python precisa entender como processar imagens. 

```python
%matplotlib inline
from PIL import Image
import requests
from io import BytesIO

# realizando um resquest em uma imagem que está em um repositório do meu GitHub
url = "https://raw.githubusercontent.com/zeolacode/Data/master/imagem.jpg"
response = requests.get(url)

# carregando/exibindo a imagem 
img = Image.open(BytesIO(response.content))

img
```
