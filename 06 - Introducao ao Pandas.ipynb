### Extraindo Features

```python
#  Gerando um segundo quadro de dados para exibir informações estatísticas 
df = df.select_dtypes(include=['int', 'float'])

# criando uma lista com o nome das colunas e outra lista para aramzenar os novos dados
headers = list(df.columns.values)
fields = []

# percorrendo as colunas e extrando as features de média, ...
for field in headers:
    fields.append({
        'name' : field, 
        'mean': df[field].mean(), # média
        'var': df[field].var(), # variância
        'sdev': df[field].std() # desvio padrão
    })

for field in fields:
    print(field)
```

```python
# Valores ausentes 
# Os valores ausentes são uma realidade do aprendizado de máquina.
# Uma prática comum é substituir os valores ausentes pelo valor mediano dessa coluna.

# verificando dados nulos em horsepower
print(f"valores NaN em horsepower? {pd.isnull(df['horsepower']).values.any()}")


# substituindo pela média!
print("Preenchendo valores ausentes...")
med = df['horsepower'].median()
df['horsepower'] = df['horsepower'].fillna(med)
# df = df.dropna() # você também pode simplesmente descartar os valores de NA

# verificando após a substituição                 
print(f"valores NaN em horsepower? {pd.isnull(df['horsepower']).values.any()}")
```

### Lidando com Outliers
- Outliers são valores que são excepcionalmente altos ou baixos.Às vezes, os outliers são simplesmente erros; isso é resultado de um erro de observação.

```python
# Remove todas as linhas em que a coluna especificada que tiver "sd" desvios padrão acima ou abaixo da média
def remove_outliers(df, name, sd):
    drop_rows = df.index[(np.abs(df[name] - df[name].mean())
                          >= (sd * df[name].std()))]
    df.drop(drop_rows, axis=0, inplace=True)
```

```python
# O código abaixo removerá todas as linhas do conjunto de dados Auto MPG em que a potência estiver mais de dois 
# desvios padrão acima ou abaixo da média


df = pd.read_csv(
    "https://data.heatonresearch.com/data/t81-558/auto-mpg.csv",
    na_values=['NA','?'])

# criar vetor de recurso
med = df['horsepower'].median()
df['horsepower'] = df['horsepower'].fillna(med)

# remove a coluna 'name'
df.drop('name',1,inplace=True)

# Elimine valores atípicos em 'horsepower'
print("Length before MPG outliers dropped: {}".format(len(df)))
remove_outliers(df,'mpg',2)
print("Length after MPG outliers dropped: {}".format(len(df)))

pd.set_option('display.max_columns', 0)
pd.set_option('display.max_rows', 5)
display(df)
```

### Codificando Valores Categóricos como Dummies

```python
# Target Encoding

# este tipo de codificação pode aumentar o poder preditivo de um modelo de aprendizado de máquina. No entanto, também aumenta drasticamente o risco de overfitting.
# Criando um pequeno dataset!

import pandas as pd
import numpy as np

np.random.seed(43)
df = pd.DataFrame({
    'cont_9': np.random.rand(10)*100,
    'cat_0': ['dog'] * 5 + ['cat'] * 5,
    'cat_1': ['wolf'] * 9 + ['tiger'] * 1,
    'y': [1, 0, 1, 1, 1, 1, 0, 0, 0, 0]
})

df
```

```python
# Em vez de criar variáveis ​​fictícias para "cachorro" e "gato", gostaríamos de alterá-lo para um número. Poderíamos usar 0 para gato, 1 para cachorro. 
# No entanto, podemos codificar mais informações do que apenas isso. O simples 0 ou 1 também funcionaria apenas para um animal. 
# Considere qual é o valor-alvo médio para gato e cachorro.

means0 = df.groupby('cat_0')['y'].mean().to_dict()
means0
```

```python
""" O perigo é que agora estamos usando o valor alvo para treinamento. Esta técnica irá potencialmente levar a overfitting. 
    A possibilidade de overfitting é ainda maior se houver um pequeno número de uma determinada categoria. 
    Para evitar que isso aconteça, usamos um fator de ponderação. Quanto mais forte o peso, mais do que categorias com um pequeno 
    número de valores tenderão para a média geral de y. Você pode realizar esse cálculo da seguinte maneira.  """

df['y'].mean()
```

```python
def calc_smooth_mean(df1, df2, cat_name, target, weight):
    # Calcular a média global
    mean = df[target].mean()

    # Calcular o número de valores e a média de cada grupo
    agg = df.groupby(cat_name)[target].agg(['count', 'mean'])
    counts = agg['count']
    means = agg['mean']

    # Calcular os meios "suavizados"
    smooth = (counts * means + weight * mean) / (counts + weight)

    # Substitua cada valor pela média suavizada de acordo
    if df2 is None:
        return df1[cat_name].map(smooth)
    else:
        return df1[cat_name].map(smooth),df2[cat_name].map(smooth.to_dict())
```

```python
# O código a seguir codifica essas duas categorias.

WEIGHT = 5
df['cat_0_enc'] = calc_smooth_mean(df1=df, df2=None, 
    cat_name='cat_0', target='y', weight=WEIGHT)
df['cat_1_enc'] = calc_smooth_mean(df1=df, df2=None, 
    cat_name='cat_1', target='y', weight=WEIGHT)

df
```
