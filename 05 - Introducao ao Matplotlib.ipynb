## Intro

```python
# importando pacote matplotlib
import matplotlib.pyplot as plt
# importando pacote NumPy
import numpy as np
```

```python
# 01 - exemplo 
"""O Matplotlib representa seus dados em gráficos, cada um dos quais pode conter um ou mais eixos, 
   uma área onde os pontos podem ser especificados em termos de coordenadas xy. A maneira mais simples 
   de criar uma Figura com um Axes é usando pyplot.subplots. 
   Podemos então usar Axes.plot para desenhar alguns dados nos Eixos:"""

arr1 = np.array([1,2,3,4,5])
arr2 = np.array([1,2,7,4,5])

# Criar uma figura contendo um único eixo:
fig, ax = plt.subplots()
# plotando arr1xarr2
ax.plot(arr1,arr2)
```

```python
print(type(ax))
print(type(fig))
```

```python
# uma figura vazia sem eixos
fig = plt.figure()
```

```python
# 02 - exemplo 

plt.plot(arr1,arr2, 'ro') # pontos vermelhos 
plt.axis([0, 6, -4, 14])  # editando axis
plt.show()
```

```python
# tempo amostrado uniformemente em intervalos de 300 ms
t = np.arange(0., 5., 0.3)

# plot com traços vermelhos, quadrados azuis e triângulos verdes
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^',)

# colocando título, nome do eixo e legenda
plt.title('SEU TÍTULO LINDO') # adicionando o título
plt.xlabel('NOME DO EIXO X') # definindo nome do eixo X
plt.ylabel('NOME DO EIXO Y') # definindo nome do eixo Y
plt.legend(labels=['t', 't^2', 't^3']) # definindo legendas
plt.grid(True) # definindo grid 

plt.show()
```

```python
# Colocando anotação no seu gráfico
fig, ax = plt.subplots()
ax.bar(arr1, arr2)
ax.annotate("Maior valor", # texto
            xy = (3, 7), 
            xytext = (3.5, 8.78), # posição do texto
            arrowprops = dict(arrowstyle="->",connectionstyle="arc3") # Arrow
            )
ax.set_ylim(0, 10) # limites do eixo x
plt.show()
```

```python
# Criando dados aleatorios 
rng = np.arange(50)
rnd = np.random.randint(0, 10, size=(3, rng.size))
yrs = 1950 + rng

# Criando figura
fig, ax = plt.subplots(figsize=(8, 5))
ax.stackplot(yrs, (rng + rnd), labels=['Eastasia', 'Eurasia', 'Oceania'])

# Título, legenda, (X,Y) label,  
ax.set_title('Crescimento combinado da dívida ao longo do tempo')
ax.legend(loc='upper left')
ax.set_ylabel('Dívida total')
ax.set_xlabel('Ano')
ax.set_xlim(xmin=yrs[0], xmax=yrs[-1]) # Definir os limites de visualização do eixo x
fig.tight_layout() # aplica-se ao objeto Figura como um todo! Ele limpa o preenchimento de espaços em branco na figura!

##########################
# Salvando figura!
##########################
fig.savefig("savefig.png") # salvando figura no formato png! 
```

```python
# Exemplo de subplots

# importando funções
from numpy import e, pi, sin, exp, cos

# criando funções 
def f(t):
    return exp(-t) * cos(2*pi*t)

def fp(t):
    return -2*pi * exp(-t) * sin(2*pi*t) - e**(-t)*cos(2*pi*t)

def g(t):
    return sin(t) * cos(1/(t+0.1))

def g(t):
    return sin(t) * cos(1/(t))


# Tamanho da figura geral
fig = plt.figure(figsize=(6, 4))

# Eixo x 
t = np.arange(-5.0, 1.0, 0.1)

# subplot 1
sub1 = fig.add_subplot(221) 
sub1.set_title('The function f') 
sub1.plot(t, f(t)) 

# subplot 2
sub2 = fig.add_subplot(222, facecolor="lightgrey")
sub2.set_title('fp, the derivation of f')
sub2.plot(t, fp(t))

# subplot 3
t = np.arange(-3.0, 2.0, 0.02) # modificando o intervalo do eixo x
sub3 = fig.add_subplot(223)
sub3.set_title('The function g')
sub3.plot(t, g(t))

# subplot 4
t = np.arange(-0.2, 0.2, 0.001) # modificando o intervalo do eixo x
sub4 = fig.add_subplot(224, facecolor="lightgrey")
sub4.set_title('A closer look at g')
sub4.set_xticks([-0.2, -0.1, 0, 0.1, 0.2])  # setando valores do eixo x   
sub4.set_yticks([-0.15, -0.1, 0, 0.1, 0.15])  # setando valores do eixo y
sub4.plot(t, g(t))

plt.plot(t, g(t))

plt.tight_layout()
plt.show()
```

```python
# Plotando 2 Axes

# Criando dados aleatorios 
x = np.random.randint(low=1, high=11, size=50)
y = x + np.random.randint(1, 5, size=x.size)
data = np.column_stack((x, y))

# Criando figura com 2 axes
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

# ax 1 
ax1.scatter(x=x, y=y, marker='o', c='r', edgecolor='b')
ax1.set_title('Scatter: $x$ versus $y$')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$') 

# ax2
ax2.hist(data, bins=np.arange(data.min(), data.max()), label=('x', 'y'))
ax2.legend(loc=(0.65, 0.8)) # posição da leganda
ax2.set_title('Frequencies of $x$ and $y$') # titulo
ax2.yaxis.tick_right() # coloca yaxis a direita
```

### Esquemas de interpolação de matriz

 - A interpolação calcula qual a cor ou valor de um pixel "deveria" ser, de acordo com diferentes esquemas matemáticos. Isso acontece é quando você redimensiona uma imagem. O número de pixels muda, mas você deseja as mesmas informações. Como os pixels são discretos, falta espaço. A interpolação é como você preenche esse espaço. É por isso que suas imagens às vezes parecem pixeladas quando você as amplia. O efeito é mais pronunciado quando a diferença entre a imagem original e a imagem expandida é maior. Vamos pegar nossa imagem e reduzi-la. Estamos efetivamente descartando pixels, mantendo apenas alguns selecionados. Agora, quando plotamos, esses dados são ampliados para o tamanho da tela. Os pixels antigos não estão mais lá, e o computador precisa desenhar pixels para preencher esse espaço.

- Usaremos a biblioteca Pillow que usamos para carregar a imagem também para redimensionar a imagem.

```python
img = Image.open('/content/teste.png')
img.thumbnail((64, 64), Image.ANTIALIAS)  # redimensiona a imagem no local
imgplot = plt.imshow(img)
# Aqui temos a interpolação padrão, bilinear, já que não demos a imshow() nenhum argumento de interpolação.
```

```python
imgplot = plt.imshow(img, interpolation="nearest")
# Aqui temos a interpolação nearest
```

```python
imgplot = plt.imshow(img, interpolation="bicubic")
# Aqui temos a interpolação bicubic
```
